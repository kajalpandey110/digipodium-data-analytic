```python
import pandas as pd
import plotly.express as px
```

```python
iris=px.data.iris()
iris
```

```python
px.scatter(iris,x='sepal_length',y='sepal_width',color='species',marginal_y='violin',marginal_x='box',trendline='ols',template='plotly_dark')
```

```python
px.scatter(iris,x='petal_length',y='petal_width',color='species',marginal_y='violin',marginal_x='box',template='plotly_dark')
```

```python
df=px.data.iris()
df['e']=df['sepal_width']/100
fig=px.scatter(df,x='sepal_width',y='sepal_length',color='species',error_x='e',error_y='e',template='plotly_dark')
fig.show()
```
